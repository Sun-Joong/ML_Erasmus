```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import Adam
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
```

```python
data  = pd.read_csv("data/creditcardfraud_shuffled.csv").values

start_train = 1
end_train = int(0.8 * len(data))
start_test = end_train + 1
end_test = len(data)

end_val = end_train
end_train = int(end_train * 0.8)
start_val = end_train + 1

train = data[start_train:end_train]
val = data[start_val:end_val]
test = data[start_test:end_test]


mean = train.mean(axis = 1)
sd = train.std(axis = 1)

train[28] = (train[28] - mean[28])/sd[28]
val[28] = (val[28] - mean[28])/sd[28]
test[28] = (test[28] - mean[28])/sd[28]

x_train = train[:, :29]
y_train = train[:, 29:]
x_val = val[:, :29]
y_val = val[:, 29:]
x_test = test[:, :29]
y_test = test[:, 29:]
```

```python
from keras import backend as K
import tensorflow as tf

K.clear_session()
config = tf.ConfigProto(intra_op_parallelism_threads=1, 
                        inter_op_parallelism_threads=4,
                        device_count = {'CPU': 1}
                       )
session = tf.Session(config=config)
K.set_session(session)
```

```python
start = time.time()
start_work = time.perf_counter() 

model = Sequential(
    [
        Dense(4, input_shape=(29,)),
        Activation('relu'),
        Dense(1),
        Activation('sigmoid'),
    ])

model

model.compile(
    optimizer=Adam(0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_train, 
    y_train,
    validation_data = (x_val, y_val),
    epochs=250,
    batch_size=512
)
end = time.time()
print(end - start)
end_work = time.perf_counter() 
print(end_work - start_work)
# Plot training & validation accuracy values
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
