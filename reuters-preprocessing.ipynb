```python
import csv
from operator import itemgetter

import nltk
from nltk.corpus import reuters
```

```python
nltk.download('reuters')
```

```python
len([_ for _ in reuters.fileids() if 'training/' in _])
```

```python
reuters.fileids('acq')
```

```python
counts = [(_, len(reuters.fileids(_))) for _ in reuters.categories()]
top_10_categories = sorted(counts, key=itemgetter(1), reverse=True)[:10]
top_10_categories = [_[0] for _ in top_10_categories]
top_10_categories
```

```python
with open('Reuters-R10-training.csv', 'wt') as stream:
    writer = csv.writer(stream)
    for category in top_10_categories:
        files = [_ for _ in reuters.fileids(category) if 'training/' in _]
        print(category, len(files))
        for _file in files:
            writer.writerow([category, ' '.join(reuters.words(_file))])
```

```python
with open('Reuters-R10-test.csv', 'wt') as stream:
    writer = csv.writer(stream)
    for category in top_10_categories:
        files = [_ for _ in reuters.fileids(category) if 'test/' in _]
        print(category, len(files))
        for _file in files:
            writer.writerow([category, ' '.join(reuters.words(_file))])
```

```python
with open('Reuters-R10-training.csv') as stream:
    reader = csv.reader(stream)
    for row in reader:
        assert(len(row) == 2)
```

```python
with open('Reuters-R10-test.csv') as stream:
    reader = csv.reader(stream)
    for row in reader:
        assert(len(row) == 2)
```
