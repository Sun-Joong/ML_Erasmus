```python
import keras
from keras.datasets import mnist
```

```python
X, y = mnist.load_data()[0]
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'retina'
```

```python
_, axes = plt.subplots(nrows=1, ncols=5, figsize=(18, 10))
for ax, image, label in zip(axes[:5], X, y):
    ax.imshow(image, cmap='gray')
    ax.set_title('Label: {}'.format(label), fontsize=20, y=1.05)
    ax.axis('off')

plt.savefig('../erasmus/images/mnist.png', dpi=150, bbox_inches='tight')
```
