```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/creditcard.csv')
df.head()
```

```python
df = df.drop(columns=['Time'])
df.head()
```

```python
df['Amount'] = (df['Amount'] - np.mean(df['Amount'])) / np.std(df['Amount'])
df.head()
```

```python
df.Class = df.Class.astype('category')
df.head()
```

```python
df['Class'].value_counts()
```

```python
492 / 284000
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0, ratio=0.05)

x = df.loc[:,df.columns != 'Class']
y = df.loc[:,'Class']

x_train, y_train = smote.fit_sample(x, y)
x_train.shape
```

```python
(y_train == 1).sum()
```

```python
from imblearn.under_sampling import RandomUnderSampler

sampler = RandomUnderSampler(random_state=0)
x_train, y_train = sampler.fit_resample(x_train, y_train)
x_train.shape
```

## Add noise

```python
mu, sigma = 0, 2.5
noise = np.random.normal(mu, sigma, x_train.shape)
x_train += noise
```

```python
(y_train == 0).sum(), (y_train == 1).sum()
```

## Some tests

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

tree = DecisionTreeClassifier(random_state=0)

cross_val_score(tree, x_train, y_train, cv = 10, scoring='accuracy')
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state=0, n_estimators=10)
cross_val_score(rf, x_train, y_train, cv=10, scoring='accuracy')
```

```python
x_train.shape
```

```python
df = pd.DataFrame.from_records(x_train)
df['label'] = y_train
df.head()
```

```python
df.to_csv('data/creditcardfraud.csv', index=False)
```

## Final export

```python
import pandas as pd
import numpy as np

np.random.seed(0)
```

```python
df = pd.read_csv('data/creditcard.csv')
df.head()
```

```python
df = df.drop(columns=['Time'])
df.head()
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=0, ratio=0.05)

x = df.loc[:,df.columns != 'Class']
y = df.loc[:,'Class']

x_train, y_train = smote.fit_sample(x, y)
x_train.shape
```

```python
from imblearn.under_sampling import RandomUnderSampler

sampler = RandomUnderSampler(random_state=0)
x_train, y_train = sampler.fit_resample(x_train, y_train)
x_train.shape
```

```python
x_train[0]
```

```python
mu, sigma = 0, 2.5
noise = np.random.normal(mu, sigma, (x_train.shape[0], x_train.shape[1] - 1))
x_train[:,:-1] += noise
```

```python
x_train[0]
```

```python
df = pd.DataFrame.from_records(x_train)
df['label'] = y_train
df.columns = ['PCA{}'.format(_) for _ in range(1, 29)] + ['amount', 'label']
df.head()
df.to_csv('data/creditcardfraud.csv', index=False)
```
